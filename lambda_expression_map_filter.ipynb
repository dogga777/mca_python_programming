```python
class Student:
    def __init__(self, name, marks=[]):
        self.name = name
        self.marks = marks

    def total_marks(self):
        return sum(self.marks)
    
    def __gt__(self, other):
        if isinstance(other, Student):
            return self.total_marks() > other.total_marks()
        elif isinstance(other, int):
            return self.total_marks() > other
    
    def __repr__(self):
        return f'Student({self.name}, {self.total_marks()})'
    

# Function to check pass/fail
def is_passed(student):
    return student.total_marks() > 200


# List of students
students = [
    Student('Kelsier', [23, 50, 67, 20, 10]),
    Student('Vin', [43, 86, 77]),
    Student('Rashek', [83, 86, 87]),
    Student('Sazed', [53, 86, 67]),
]

# Filtering passed students
passed_students = list(filter(is_passed, students))

print(passed_students)
```
