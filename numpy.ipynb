```python
x = np.random.randint(1,100,(10,10))
print(x)
```

```python
print("first row",x[0])
print("last row", x[-1])
print("second row", x[1])
print('first 5 element of first row', x[0, :5])
print('first 3 values of last row', x[0, -3:])
print('first 3 elements of last row', x[-1, :3])
print('third row, all except first 2 and last 2', x[2, 2:-2])
print('top 3 rows\n', x[:3])
print('last 3 rows\n', x[-3:])
print("all rows except the first and last\n", x[1:-1])
```

```python
print("last column of x", x[:, 0])
print("last column of x", x[:, -1])
print("5th column of x", x[:, 4])
print("first 3 columns\n", x[:, :3])
print("last  4 columns\n", x[:, -4:])
```

```python
print('all the odd columns', x[:, 1::2])
```

```python
# boolean indexing aka filtering
x = np.random.randint(1,10, (4,10))
print(x)
```

```python
x>3  # boolean filter - a matrix of True and False, with same dimension as x matrix
```

```python
x[x>3]  # using boolean filter to extract values that are true
```

```python
filter1 = x > 3
filter2 = x < 7
# extract using more than 1 filter
x[filter1 & filter2] # or x[(x>3) & (x<7)]
```

### USEFUL METHODS AND FUNCTIONS    

```python
# np.arange(start, stop, step)
z = np.arange(-10, 10, 3)
print(z)
```

```python
z2 = np.arange(-10, 10, .3)
print(z2)
print(z2.size)
```

```python
# np.linspace(start, stop, numofelement) => start and stop are inclusive
z3 = np.linspace(2,5, 25)
print(z3)
```

```python
# np.reshape(shape)
import numpy as np

z3 = np.arange(15)  # creates [0, 1, 2, ..., 14]

print(z3.size)  # output: 15
print("3 rows, 5 columns")
print(z3.reshape((3,5)))
print("5 rows, 3 columns")
print(z3.reshape((5,3)))
```

```python
x = np.linspace(1,100,30, dtype=int).reshape((5,3,2))
print(x)
```

```python
t = np.arange(-3, 3, .05)
x = np.sin(t)
y = np.cos(t)
print(t)
print(x)
print(y)
```

```python
import matplotlib.pyplot as plt
plt.plot(t,x)
plt.plot(t,y)
plt.show()
```
