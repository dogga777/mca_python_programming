Task 1: Create a Rectangle Class

Create a class Rectangle with methods to set length and width, and calculate area and perimeter.

Hint: Use instance variables self.length and self.width. Area = length √ó width, Perimeter = 2 √ó (length + width).

```python
class Rectangle:
    def __init__(self, length=0, width=0):
        self.length = length
        self.width = width
        
    def set_length(self, length):
        self.length = length
        
    def set_width(self, width):
        self.width = width
        
    def area(self):
        return self.length * self.width
    
    def perimeter(self):
        return 2 * (self.length + self.width)
    
rect = Rectangle()
rect.set_length(10)
rect.set_width(5)

print("Length:", rect.length)
print("Width:", rect.width)
print("Area:", rect.area())
print("Perimeter:", rect.perimeter())


# __init__ ‚Üí Constructor that initializes length and width (default 0).
# set_length() and set_width() ‚Üí Used to assign values.
# area() ‚Üí Returns length √ó width.
# perimeter() ‚Üí Returns 2 √ó (length + width).
```

Task 2: Student Grade Manager

Create a class Student that stores a student‚Äôs name and three subject marks. Add methods to calculate the total marks and average marks.

Hint: Use instance variables for name and marks. Total = sum of all marks, Average = total / number of subjects.

```python
class Student:
    def __init__(self, name, mark1, mark2, mark3):
        self.name = name
        self.mark1 = mark1
        self.mark2 = mark2
        self.mark3 = mark3
        
    def total_marks(self):
        return self.mark1 + self.mark2 + self.mark3
    
    def average_marks(self):
        return self.total_marks() / 3
    
    
Student1 = Student("Chandra", 85, 90, 80)

print("Student Name:", Student1.name)
print("Marks:", Student1.mark1, Student1.mark2, Student1.mark3)
print("Total Marks:", Student1.total_marks())
print("Average Marks:", Student1.average_marks())

# __init__() ‚Üí Initializes name and 3 subject marks.
# total_marks() ‚Üí Returns the sum of all marks.
# average_marks() ‚Üí Returns the average by dividing total by 3.
```

Task 3: Temperature Converter Create a class Temperature with methods to set temperature in Celsius and convert it to Fahrenheit and Kelvin. Hint: Fahrenheit = (Celsius √ó 9/5) + 32, Kelvin = Celsius + 273.15.

```python
class Temperature:
    def __init__(self, celsius=0):
        self.celsius = celsius
        
    def set_celsius(self, celsius):
        self.celsius = celsius
        
    def to_fahrenheit(self):
        return(self.celsius* 9/5) + 32
    
    def to_kelvin(self):
        return self.celsius + 273.15
    
temp = Temperature()
temp.set_celsius(25)

print("Temperature in celsius:", temp.celsius)
print("Temperature in Fahrenheit:", temp.to_fahrenheit())
print("Temperature in Kelvin:", temp.to_kelvin)


# __init__() ‚Üí Initializes the Celsius value (default = 0).
# set_celsius() ‚Üí Used to assign a temperature value in Celsius.
# to_fahrenheit() ‚Üí Converts Celsius to Fahrenheit using the formula (C √ó 9/5) + 32.
# to_kelvin() ‚Üí Converts Celsius to Kelvin using C + 273.15.
```

Task 4: Shopping Cart Create a class ShoppingCart with methods to add items (with prices), remove items, and calculate the total price. Hint: Use a list to store items and their prices. You can use a list of dictionaries like [{'name': 'item1', 'price': 100}].

```python

class ShoppingCart:
    
    def __init__(self):
        self.items = []   

    
    def add_item(self, name, price):
        self.items.append({'name': name, 'price': price})
        print(f"‚úÖ '{name}' added to cart at ‚Çπ{price}")

    
    def remove_item(self, name):
        for item in self.items:
            if item['name'] == name:
                self.items.remove(item)
                print(f"üóëÔ∏è '{name}' removed from cart")
                return
        print(f"‚ö†Ô∏è Item '{name}' not found in cart!")

    
    def total_price(self):
        return sum(item['price'] for item in self.items)

    
    def show_cart(self):
        if not self.items:
            print("üõí Your cart is empty.")
        else:
            print("\nItems in your cart:")
            for item in self.items:
                print(f"- {item['name']}: ‚Çπ{item['price']}")
            print(f"\nüí∞ Total Price: ‚Çπ{self.total_price()}")



cart = ShoppingCart()
cart.add_item("Laptop", 55000)
cart.add_item("Mouse", 800)
cart.add_item("Keyboard", 1500)

cart.show_cart()        
cart.remove_item("Mouse")
cart.show_cart()        
```

Task 5: Employee Management

Create a class Employee with methods to set employee details (name, ID, salary) and give a raise (increase salary by a given percentage).

Hint: Store name, ID, and salary as instance variables. For a 10% raise: new_salary = old_salary * 1.10.

```python

class Employee:
    
    def __init__(self, name="", emp_id=0, salary=0.0):
        self.name = name
        self.emp_id = emp_id
        self.salary = salary

    
    def set_details(self, name, emp_id, salary):
        self.name = name
        self.emp_id = emp_id
        self.salary = salary

    
    def display_details(self):
        print(f"\nEmployee Details:")
        print(f"Name: {self.name}")
        print(f"Employee ID: {self.emp_id}")
        print(f"Salary: ‚Çπ{self.salary:.2f}")

    
    def give_raise(self, percentage):
        increase = self.salary * (percentage / 100)
        self.salary += increase
        print(f"\nüéâ Salary increased by {percentage}%")
        print(f"New Salary: ‚Çπ{self.salary:.2f}")



emp = Employee()  
emp.set_details("Anjali", 101, 50000)
emp.display_details()

emp.give_raise(10)  
emp.display_details()

```
